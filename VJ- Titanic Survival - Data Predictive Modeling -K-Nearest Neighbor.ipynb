```python
Titanic_data_Cleaned = Titanic_data.copy()
```

```python
data = [Titanic_data_Cleaned]
```

```python
ports = {"S": 0, "C": 1, "Q": 2} # Convert embarked
for dataset in data:
    dataset['embarked'] = dataset['embarked'].map(ports)
```

```python
# Remove all NULLS in the Embarked column
for dataset in data:
    dataset['embarked'] = dataset['embarked'].fillna('S');
   
```

```python
data
```

```python
genders = {"male": 0, "female": 1}  # Convert Sex
for dataset in data:
    dataset['sex'] = dataset['sex'].map(genders)
```

```python
# Remove all NULLS in the Fare column
for dataset in data:
    dataset['fare'] = dataset['fare'].fillna(Titanic_data_Cleaned['fare'].median())
```

```python
for dataset in data: # Mapping Age
    dataset.loc[ dataset['fare'] <= 7.91, 'fare']  = 0
    dataset.loc[(dataset['fare'] > 7.91) & (dataset['fare'] <= 14.454), 'fare'] = 1
    dataset.loc[(dataset['fare'] > 14.454) & (dataset['fare'] <= 31), 'fare']   = 2
    dataset.loc[ dataset['fare'] > 31, 'Fare']       = 3
   
```

```python
 dataset['fare'] = dataset['fare'].astype(int)
```

```python
# Remove all NULLS in the Age column
for dataset in data:
    age_avg = dataset['age'].mean()
    age_std = dataset['age'].std()
    age_null_count = dataset['age'].isnull().sum()
    age_null_random_list = np.random.randint(age_avg - age_std, age_avg + age_std, size=age_null_count)
   # Next line has been improved to avoid warning
    dataset.loc[np.isnan(dataset['age']), 'age'] = age_null_random_list
    dataset['age'] = dataset['age'].astype(int)
```

```python
titanic_train = Titanic_data_Cleaned.drop(['name','home.dest','cabin','ticket','Row', 'Fare','parch', 'sibsp','embarked'], axis=1) # drop missing values
```

```python
# Modeling the Data - KNN Algorithm

#Let's use knn algorithm on numeric columns
features = ["pclass", "sex",  "age", "fare"]
X_train = titanic_train[features]
y_train = titanic_train["survived"]
X_test = titanic_train[features]
y_test = titanic_train["survived"]

print("Training split input- ", X_train.shape)
print("Testing split input- ", X_test.shape)
```

```python
# K-NN classification with K=2. 

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train, y_train)
KNNpredictions = knn.predict(X_test)
```

```python
print(classification_report(y_test, KNNpredictions))
print('Accuracy of K-Nearest Neighbor on Titanic Data: {:.2f}'.format(round(knn.score(X_test, y_test) * 100, 2)));
```
